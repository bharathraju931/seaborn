```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
avg_total_bill = tips.groupby(['day','time'])['total_bill'].mean().reset_index()
avg_total_bill

max_tip = tips.groupby('smoker')['tip'].max().reset_index()
max_tip


plt.figure(figsize=(8,5))
sns.barplot(data=avg_total_bill,x='day',y='total_bill', hue = 'time')
plt.title("average total bill by day and time")
plt.ylabel("average total bill")
plt.xlabel("day")
plt.legend(title = "time")
plt.show()
plt.figure(figsize = (5,4))
sns.barplot(data = max_tip,x='smoker',y='tip')
plt.title("maximum tip by smoker")
plt.ylabel("max_tip")
plt.xlabel("smoker")
plt.show()

```
